```python
import pandas as pd
h = pd.read_excel('한국어 학습용 어휘 목록.xls')

# 순위,등급,풀이 지우기
h = h.drop(['순위', '등급', '풀이'], axis = 1)

# 품사가 명사, 고유명사인 단어들만 남기기
nh = h[(h['품사'] == '명') | (h['품사'] == '고')]

s = set()
result = ''
for word in nh['단어']:
    for i in range(len(word)):
        if word[i].isdigit():
            result = word[:i]
            break
    s.add(result) # 중복제거

# 잘 들어갔는지 파악
s = list(s)
print(s[0])
print(s[-1])

# txt 파일로 내보내기
fw = open('한국어 어휘 목록.txt','w')
for i in range(1,len(s)):
    fw.write(s[i])
    fw.write('\n')
fw.close()
```
